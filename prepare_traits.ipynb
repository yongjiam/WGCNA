```python
import pandas as pd
import numpy as np
from scipy.stats import zscore
```

```python
df=pd.read_excel('meta_traits.xlsx',sheet_name='processed',header=0)
df.head()
```

```python
# Define a function to remove outliers based on Z-scores
def remove_outliers(values, z_threshold=3):
    values = np.array(values, dtype=float)  # Convert to numpy array with float data type
    z_scores = zscore(values)
    mask = np.abs(z_scores) < z_threshold
    return values[mask]
```

```python
df.dtypes
```

```python
# Iterate through each row and calculate means after removing outliers
for index, row in df.iterrows():
    replicates = row[-6:]  # Extract the last six columns
    replicates_cleaned = remove_outliers(replicates)
    mean_value = np.mean(replicates_cleaned)
    df.at[index, 'Mean'] = mean_value
```

```python
df.head()
```

```python
pivoted_df = df.pivot(index='sample', columns='meta', values='Mean')
pivoted_df.head()
```

```python
pivoted_df = pivoted_df.reset_index()
pivoted_df.head()
```

```python
# Define a dictionary with replacement mappings
replacement_dict = {
    'WT-': 'WT_',
    'hvhggt-': 'hggt_'
}
for old_str, new_str in replacement_dict.items():
    pivoted_df['sample'] = pivoted_df['sample'].str.replace(old_str, new_str)

pivoted_df.head()
```

```python
merged_df = pd.DataFrame()
suffixes = ['_1', '_2', '_3']
for suff in suffixes:
    df_tem = pivoted_df.copy()
    df_tem['sample'] = df_tem['sample'] + suff
    merged_df = pd.concat([merged_df, df_tem], ignore_index=True)
```

```python
merged_df = merged_df.sort_values(by='sample', ascending=True)
merged_df.head()
```
